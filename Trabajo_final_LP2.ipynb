```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
import requests
import pandas as pd
```

```python
#Leyendo los Gen abrev de la hoja C1 del excel
gen_abreviado1=[]
DF = pd.read_excel('Genes identificados con vecinos.xlsx', sheet_name="C1",skiprows=4)
genasociado= DF["Gen Abrev"]

for i in genasociado:
    gen_abreviado1.append(i)
print(gen_abreviado1)
len(gen_abreviado1)    # lista con nombre "gen_abreviado1" contiene los nombres de genes en la C1
```

```python
## leyendo los Gen abrev de las hojas que quedan
hojas_excel = ["C2","C3","C4-M","C5","C6","C7-J","C8-N","C9","C10","C11"]
gen_abreviado2=[]
for i in hojas_excel:
    datos = pd.read_excel('Genes identificados con vecinos.xlsx', sheet_name=i,skiprows=1)
    genabreviado=datos["Gen Abrev"]
    for nombre in genabreviado:
        gen_abreviado2.append(nombre)
print(gen_abreviado2)
len(gen_abreviado2)
```

```python
#################################################
# COLUMNA1 TODOS LOS GEN_ABREV
#################################################


total_Gen_Abrev = gen_abreviado1 + gen_abreviado2
total_Gen_Abrev
print(len(total_Gen_Abrev))
print(total_Gen_Abrev)
```

```python
#### LIMPIEZA DE DATOS

#1) PRIMERA LIMPIEZA: ELIMINAR LOS GENES QUE SE REPITEN 
data = {'Gen abrev': total_Gen_Abrev}
df = pd.DataFrame(data, columns = ['Gen abrev'])
#h es una variable que contiene solo las observaciones únicas
h=df["Gen abrev"].unique()
oficial_Gen = pd.DataFrame(h, columns = ['Gen abrev'])
## Con vale count podemos observar que no se repiten los genes
oficial_Gen["Gen abrev"].value_counts()
###
LISTA_DE_GENES =[]
for i in range(2456):
   LISTA_DE_GENES.append(oficial_Gen["Gen abrev"][i])

#print(LISTA_DE_GENES)
#print(len(LISTA_DE_GENES))
LISTA_DE_GENES.remove('TRNAC-GCA ')
print(LISTA_DE_GENES)                                                            #### LISTA_DE_GENES oficiales para el excel
```

```python
#para crear los links correspondientess de busqueda tenemos que hacer una modificaciones
lista=[]
for genes in LISTA_DE_GENES:
    lista.append(genes.replace(" ", "+"))
#print(lista)

lista1=[]
for genes in lista:
    lista1.append(genes.replace("(","%28"))
#print(lista1)

query=[]
for genes in lista1:
    query.append(genes.replace(")","%29"))
#print(lista2)    
    
```

```python
links=[]

for i in query[2200:2400]:
    links.append("https://www.uniprot.org/uniprot/?query="+ str(i) + "+Bos+taurus&sort=score")
#print(links)
```

```python
#######################
# COLUMNA2 ENTRY (C1)
#######################


import urllib.request, urllib.parse, urllib.error #(solicitud,analizar,error)
import re
import ssl  
ctx = ssl.create_default_context()                    #crear un contexto predeterminado
ctx.check_hostname = False                            #comprobar el nombre de host
ctx.verify_mode = ssl.CERT_NONE                       #modo de verificación

lista_entry = []

for codigo in links:
    url = codigo
    html = urllib.request.urlopen(url, context=ctx).read()  
    link = re.findall(b'<a href=".uniprot.([a-zA-Z].*?)"', html) 
    if len(link)==0:
        lista_entry.append("Na")
    else:
        objeto=link[0].decode()
        lista_entry=lista_entry +[objeto] 
print(lista_entry)   
len(lista_entry)
```
