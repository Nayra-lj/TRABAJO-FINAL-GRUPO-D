```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
import requests
import pandas as pd
```

```python
#Leyendo los Gen abrev de la hoja C1 del excel
gen_abreviado1=[]
DF = pd.read_excel('Genes identificados con vecinos.xlsx', sheet_name="C1",skiprows=4)
genasociado= DF["Gen Abrev"]

for i in genasociado:
    gen_abreviado1.append(i)
print(gen_abreviado1)
len(gen_abreviado1)    # lista con nombre "gen_abreviado1" contiene los nombres de genes en la C1
```

```python
## leyendo los Gen abrev de las hojas que quedan
hojas_excel = ["C2","C3","C4-M","C5","C6","C7-J","C8-N","C9","C10","C11"]
gen_abreviado2=[]
for i in hojas_excel:
    datos = pd.read_excel('Genes identificados con vecinos.xlsx', sheet_name=i,skiprows=1)
    genabreviado=datos["Gen Abrev"]
    for nombre in genabreviado:
        gen_abreviado2.append(nombre)
print(gen_abreviado2)
len(gen_abreviado2)
```

```python
#################################################
# COLUMNA1 TODOS LOS GEN_ABREV
#################################################


total_Gen_Abrev = gen_abreviado1 + gen_abreviado2
total_Gen_Abrev
print(len(total_Gen_Abrev))
print(total_Gen_Abrev)
```

```python
#### LIMPIEZA DE DATOS
from pandas import ExcelWriter 
#1) PRIMERA LIMPIEZA: ELIMINAR LOS GENES QUE SE REPITEN 
data = {'Gen abrev': total_Gen_Abrev}
df = pd.DataFrame(data, columns = ['Gen abrev'])
#h es una variable que contiene solo las observaciones únicas
h=df["Gen abrev"].unique()
oficial_Gen = pd.DataFrame(h, columns = ['Gen abrev'])
## Con vale count podemos observar que no se repiten los genes
oficial_Gen["Gen abrev"].value_counts()
###
LISTA_DE_GENES =[]
for i in range(2456):
    LISTA_DE_GENES.append(oficial_Gen["Gen abrev"][i])

#print(LISTA_DE_GENES)
#print(len(LISTA_DE_GENES))
LISTA_DE_GENES.remove('TRNAC-GCA ')
#### LISTA_DE_GENES oficiales para el excel

LISTA_GENES_FINAL=pd.DataFrame(LISTA_DE_GENES,columns=['Gen abrev'])

LISTA_GENES_FINAL


# creación del excel a partir del Dataframe

writer = ExcelWriter('./Gen_abrev.xlsx')
LISTA_GENES_FINAL.to_excel(writer, 'C1', index=True)
writer.save()
LISTA_GENES_FINAL #Una vez Limpiada la data, trabajamos oficialmente y nos repartimos el trabajo,
#Nos repartimos 818,818,819
#1roCorreria de [0:818]
#2doCorreria de [818:1636]
#3roCorreria de [1636:2455]

##########################################
#EN MI CASO ME TOCO CORRER DE [1636:2455]#
##########################################
```

```python
#Creacion querys de busqueda, tenemos que hacer unas modificaciones a la LISTA_DE_GENES
#para crear los links correspondientess de busqueda tenemos que hacer una modificaciones
lista=[]
for genes in LISTA_DE_GENES:
    lista.append(genes.replace(" ", "+"))
#print(lista)

lista1=[]
for genes in lista:
    lista1.append(genes.replace("(","%28"))
#print(lista1)

lista2=[]
for genes in lista1:
    lista2.append(genes.replace(")","%29"))
print(lista2)

query=[]
for genes in lista2:
    query.append(genes.replace("ó","%C3%B3"))
len(query)
```

```python
#Creando mis links a partir de mis querys
links=[]

for i in query[0:2455]:
    links.append("https://www.uniprot.org/uniprot/?query="+ str(i) + "+Bos+taurus&sort=score")
#Ya tengo los Links
len(links)
```

```python
#######################
# COLUMNA ENTRY (2)
#######################


import urllib.request, urllib.parse, urllib.error #(solicitud,analizar,error)
import re
import ssl  
ctx = ssl.create_default_context()                    #crear un contexto predeterminado
ctx.check_hostname = False                            #comprobar el nombre de host
ctx.verify_mode = ssl.CERT_NONE                       #modo de verificación

LISTA_ENTRY = []
#Corriendo todos los entrys, en mi caso lo hare de 100 en 100
for codigo in links[0:100]:
    url = codigo
    html = urllib.request.urlopen(url, context=ctx).read()  
    link = re.findall(b'<a href=".uniprot.([a-zA-Z].*?)"', html) 
    if len(link)==0:
        LISTA_ENTRY.append("Na")
    else:
        objeto=link[0].decode()
        LISTA_ENTRY = LISTA_ENTRY +[objeto] 
print(LISTA_ENTRY)  
```

```python
# len(LISTA_ENTRY)
#Conviertiendo a excel mis entrys encontrados
dfdf=pd.DataFrame(LISTA_ENTRY)

dfdf.to_excel("LISTA_ENTRY5.xlsx")
```

```python
############################
# COLUMNA ENTRY NAME (3)
############################

lista2 = []

for codigo in links[2336:]:
    url = codigo
    html = urllib.request.urlopen(url, context=ctx).read()  
    link = re.findall(b'<td>([A-Z].*?)<.td>', html) 
    if len(link)==0:
        lista2.append('Na')
    else:
        objeto=link[0].decode()
        lista2=lista2 +[objeto] 
print(lista2)    
len(lista2)
```

```python
#Guardando listas en excel para luego compilarlos en mi reporte
dfdf2=pd.DataFrame(lista2)

dfdf2.to_excel("Entry_name6.xlsx")
```

```python
#Guarde todos mis Entrys en una excel "BORRADOR.xlsx"
#De esta forma me facilita todos los entrys y hacerle las iteraciones en cualquier rango
borrador=pd.read_excel("BORRADOR.xlsx",header=None)
Entry=[]
#Guardo estos Entrys en una lista llada Entry
#Hago iteraciones para guardar todos los elementos de borrador en entrys 
for i in range(len(borrador)):
    Entry.append(borrador[0][i])
len(Entry)

```

```python
###########################################
# COLUMNA PROTEIN  (4)
###########################################
# Nuestro Entry posee 819 filas 
##Empezaremos corriendo los 100 primeros entrys
urls=[]
#Se realiza una iteracion for a los Entrys, lo cual devuelve urls,
for i in Entry[700:]:
    if i=="Na":
        urls.append(0)
    else:
        urls.append("https://www.uniprot.org/uniprot/"+str(i))
#print(urls)

PROTEINA=[]
#las urls son iteradas con el metodo for
for i in urls:
    #Si el url es '0', agrega un Na a mi lista 
    if i==0:
        PROTEINA.append("Na")
#print(PROTEINA)
    #caso contrario, hace un request y en base a ello decodifica mis proteinas
    else:        
        url = i
        html = urllib.request.urlopen(url, context=ctx).read()  
        links = re.findall(b'<h1 property="name">([\[*a-zA-Z0-9].*?)<', html) 
        for link in links:
            nuevo=link.decode()
            PROTEINA = PROTEINA+[nuevo]
            
print(PROTEINA)
len(PROTEINA)
```

```python
#Guardando listas en excel para luego compilarlos en mi reporte
dfdf3=pd.DataFrame(PROTEINA)

dfdf3.to_excel("proteina8.xlsx")
```

```python
########################################
# COLUMNA5 GENES  (5)
########################################
#SE REPITE EL PROCESO DE LA COLUMNA PROTEIN
urls=[]
# Nuestro Entry posee 819 filas 
##Empezaremos corriendo los 100 primeros entrys
for i in Entry[0:100]:
    if i=="Na":
        urls.append(0)
    else:
        urls.append("https://www.uniprot.org/uniprot/"+str(i))

GENES=[]

for i in urls:
    if i==0:
        GENES.append(0)
    else:        
        url = i
        html = urllib.request.urlopen(url, context=ctx).read()  
        links = re.findall(b'<h2>([A-Z].*?)<', html) 
        if len(re.findall(b'<h2>([A-Z].*?)<', html))==1:  # Algunos Entrys no tienen el nombre de Gene, por lo tanto links saldría vacío, con len =0
            for link in links:
                nuevo=link.decode()
                GENES=GENES+[nuevo]
        else:
        
            GENES=GENES+["N/A"]
#print(GENES)
len(GENES)






```

```python
#Guardando listas en excel para luego compilarlos en mi reporte
dfdf4=pd.DataFrame(GENES)

dfdf4.to_excel("genes8.xlsx")
```

```python
################################
# COLUMNA ORGANISMO  (6)
################################
#SE REPITE EL PROCESO DE LA COLUMNA PROTEIN
urls=[]
#Se realiza una iteracion for a los Entrys, lo cual devuelve urls,
for i in Entry[0:100]:
    if i=="Na":
        urls.append(0)
    else:
        urls.append("https://www.uniprot.org/uniprot/"+str(i))
        
        
ORGANISMO=[]

for i in urls:
    if i==0:
        ORGANISMO.append(0)
    else:        
        url = i
        html = urllib.request.urlopen(url, context=ctx).read()  
        links = re.findall(b'<em>([A-Z].*?)</em>', html) 
        if  len(links)==2:     # si hay más de un objeto en links o si hay exactamente 2.
            ORGANISMO=ORGANISMO+ [links[1]]   #agrega  a la lista solo el segundo objeto de la lista
        else:
            for link in links:
                nuevo=link.decode()
                ORGANISMO=ORGANISMO+[nuevo]




len(ORGANISMO)
```

```python
#Guardando listas en excel para luego compilarlos en mi reporte
dfdf6=pd.DataFrame(ORGANISMO)

dfdf6.to_excel("organismo8.xlsx")
```

```python
################################
# COLUMNA STATUS  (7)
################################
#SE REPITE EL PROCESO DE LA COLUMNA PROTEIN
urls=[]
#Se realiza una iteracion for a los Entrys, lo cual devuelve urls,
for i in Entry[0:100]:
    if i=="Na":
        urls.append(0)
    else:
        urls.append("https://www.uniprot.org/uniprot/"+str(i))
        
STATUS=[]

for i in urls:
    if i==0:
        STATUS.append(0)
    else:        
        url = i
        html = urllib.request.urlopen(url, context=ctx).read()  
        links = re.findall(b'<span>([UR].*viewed)', html) 
        for link in links:
            nuevo=link.decode()
            STATUS=STATUS+[nuevo]
#print(STATUS)

len(STATUS)
```

```python
#Guardando listas en excel para luego compilarlos en mi reporte
dfdf7=pd.DataFrame(STATUS)

dfdf7.to_excel("status8.xlsx")
```

```python
################################
# COLUMNA GO - Biological process  (8)
################################
#SE REPITE EL PROCESO DE LA COLUMNA PROTEIN
urls=[]
#Se realiza una iteracion for a los Entrys, lo cual devuelve urls,
for i in Entry[700:]:
    if i=="Na":
        urls.append(0)
    else:
        urls.append("https://www.uniprot.org/uniprot/"+str(i))




GO_Biological_process2 = []

for i in urls:
    if i==0:
        GO_Biological_process2.append("Na")
    else:
        uclient=urlopen(i)
        pagehtml=uclient.read()
        uclient.close()                                    
        pagesoup=soup(pagehtml, "html.parser")
        page=pagesoup.find("ul",{"class":"noNumbering biological_process"}) 
        
        if page==None:
            GO_Biological_process2.append("Na")
        else:
            location=page.find_all("a",{"onclick":"window.ga('UniProt-Entry-View', 'click', 'Display-GO-Term');"})
            #print(location)
            cadena = ""
            for i in range(0,len(location)):
                if location[i].text=="":
                    cadena = cadena + "-"
                else:
                    #GO_Biological_process.append(location[i].text)
                    cadena = cadena + location[i].text + "."
                #print(type(location[i].text))

            GO_Biological_process2.append(cadena)
            
len(GO_Biological_process2)
```

```python
#Guardando listas en excel para luego compilarlos en mi reporte
dfdf8=pd.DataFrame(GO_Biological_process2)

dfdf8.to_excel("gobiological8.xlsx")
```

```python
########### COLUMNA molecular function (9) ##############
#SE REPITE EL PROCESO DE LA COLUMNA PROTEIN PARA EXTRAER LOS urls
urls=[]
#Se realiza una iteracion for a los Entrys, lo cual devuelve urls,
for i in Entry[400:500]:
    if i=="Na":
        urls.append(0)
    else:
        urls.append("https://www.uniprot.org/uniprot/"+str(i))


col_molecular_f=[]

for link in urls:
    if link !=0:
        uclient=urlopen(link)
        pagehtml=uclient.read()
        uclient.close()
        pagesoup=soup(pagehtml, "html.parser")


        if pagesoup.find("div",{"id":"function"}):
            section=pagesoup.find("div",{"id":"function"})
            if section.find("table",{"class":"databaseTable"}): 
                tabla=section.find("table",{"class":"databaseTable"})
                span=tabla.find("tbody")
                td=tabla.find_all("td")
                for i in range(len(td)):
                    if td[i].text == "Molecular function":
                        span=td[i+1].find_all("span",recursive=False)
                        span_0=span[0].find_all("a",recursive=False)
                        string=""
                        for spancito in span_0:
                            string= string  + spancito.text + ","

                        col_molecular_f.append(string)    


                        break
                    else:
                        if i ==len(td)-1: #Esto pregunta si es el ultimo 'td', si es asi y no lo encuentra, agrega 'Na'
                            col_molecular_f.append("Na")
                        
                        else:
                            continue #Caso contrario, no fuese el ultimo elemento td, continua la iteración
            else :
                col_molecular_f.append("Na")
        else:
            col_molecular_f.append("Na")
    else:
        col_molecular_f.append("Na")
        
        
print(col_molecular_f)

```

```python
#Guardando listas en excel para luego compilarlos en mi reporte
dfdf9=pd.DataFrame(col_molecular_f)

dfdf9.to_excel("molecular_f1.xlsx")
```
