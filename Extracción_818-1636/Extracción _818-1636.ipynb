```python
from bs4 import BeautifulSoup as soup
import requests
import urllib.request   
import pandas as pd

from urllib.request import urlopen
from bs4 import BeautifulSoup as soup

```

```python
#Lectura de genes limpios(sin repetición)
df = pd.read_excel('Gen_abrev.xlsx',index_col=0, names=['N°','Gen abrev', 'Conteo'])
Datos = df["Gen abrev"]
datos= Datos.iloc[818:1636]
len(datos)
```

```python
lista_genes=[]
for i in range(len(datos)):
    lista_genes.append(datos.iloc[i])
lista_genes
```

```python
#para crear los links correspondientess de busqueda tenemos que hacer una modificaciones
lista=[]
for genes in lista_genes:
#len(Datos.iloc[818:1637])
    lista.append(genes.replace(" ", "+"))
#print(lista)

lista1=[]
for genes in lista:
    lista1.append(genes.replace("(","%28"))
#print(lista1)

query=[]
for genes in lista1:
    query.append(genes.replace(")","%29"))
print(query)
len(query)
```

```python
LISTA_GENES= query[200:400]
print(LISTA_GENES)
len(LISTA_GENES)
```

```python
links=[]

for i in query:
    links.append("https://www.uniprot.org/uniprot/?query="+ str(i) + "+Bos+taurus&sort=score")
#print(links)
```

```python
len(links)
```

### SEGUNDA EXTRACCIÓN DE LOS 200 siguientes LINKS

```python
#######################
# COLUMNA2 ENTRY 
#######################


import urllib.request, urllib.parse, urllib.error #(solicitud,analizar,error)
import re
import ssl  
ctx = ssl.create_default_context()                    #crear un contexto predeterminado
ctx.check_hostname = False                            #comprobar el nombre de host
ctx.verify_mode = ssl.CERT_NONE                       #modo de verificación

LISTA_ENTRY = []

for codigo in links[200:400]:
    url = codigo
    html = urllib.request.urlopen(url, context=ctx).read()  
    link = re.findall(b'<a href=".uniprot.([a-zA-Z].*?)"', html) 
    if len(link)==0:
        LISTA_ENTRY.append("Na")
    else:
        objeto=link[0].decode()
        LISTA_ENTRY = LISTA_ENTRY +[objeto] 
print(LISTA_ENTRY)   
len(LISTA_ENTRY)
```

```python
############################
# COLUMNA3 ENTRY NAME 
############################

LISTA_ENTRYNAME = []

for codigo in links[200:400]:
    url = codigo
    html = urllib.request.urlopen(url, context=ctx).read()  
    link = re.findall(b'<td>([A-Z].*?)<.td>', html) 
    if len(link)==0:
        LISTA_ENTRYNAME.append('Na')
    else:
        objeto=link[0].decode()
        LISTA_ENTRYNAME=LISTA_ENTRYNAME +[objeto] 
print(LISTA_ENTRYNAME)    
len(LISTA_ENTRYNAME)
```

```python
###########################################
# COLUMNA4 PROTEIN  
###########################################

urls=[]

for i in LISTA_ENTRY:
    if i=="Na":
        urls.append(0)
    else:
        urls.append("https://www.uniprot.org/uniprot/"+str(i))
#print(urls)

PROTEINA=[]

for i in urls:
    if i==0:
        PROTEINA.append("Na")
#print(PROTEINA)
    else:        
        url = i
        html = urllib.request.urlopen(url, context=ctx).read()  
        links = re.findall(b'<h1 property="name">([a-zA-Z0-9].*?)<', html) 
        for link in links:
            nuevo=link.decode()
            PROTEINA = PROTEINA+[nuevo]
            
print(PROTEINA)
len(PROTEINA)
```

```python
########################################
# COLUMNA5 GENES 
########################################

GENES=[]

for i in urls:
    if i==0:
        GENES.append("Na")
    else:        
        url = i
        html = urllib.request.urlopen(url, context=ctx).read()  
        links = re.findall(b'<h2>([A-Z].*?)<', html) 
        #print(links)
        if len(re.findall(b'<h2>([A-Z].*?)<', html))==1:  # Algunos Entrys no tienen el nombre de Gene, por lo tanto links saldría vacío, con len =0
            for link in links:
                nuevo=link.decode()
                GENES=GENES+[nuevo]
        else:
        
            GENES=GENES+["N/A"]
print(GENES)
len(GENES)
```

```python
len(urls)
```

```python
################################
# COLUMNA6 ORGANISMO 
################################

ORGANISMO=[]

for i in urls:
    if i==0:
        ORGANISMO.append("Na")
    else:        
        url = i
        html = urllib.request.urlopen(url, context=ctx).read()  
        links = re.findall(b'<em>([A-Z].*?)</em>', html) 
        if  len(links)==2:     # si hay más de un objeto en links o si hay exactamente 2.
            ORGANISMO=ORGANISMO+ [links[1]]   #agrega  a la lista solo el segundo objeto de la lista
        else:
            for link in links:
                nuevo=link.decode()
                ORGANISMO=ORGANISMO+[nuevo]
print(ORGANISMO)
len(ORGANISMO)
```

```python
################################
# COLUMNA7 STATUS  
################################

STATUS=[]

for i in urls:
    if i==0:
        STATUS.append("Na")
    else:        
        url = i
        html = urllib.request.urlopen(url, context=ctx).read()  
        links = re.findall(b'<span>([UR].*viewed)', html) 
        for link in links:
            nuevo=link.decode()
            STATUS=STATUS+[nuevo]
print(STATUS)
len(STATUS)
```

```python
################################
# COLUMNA8 GO - Biological process  
################################

GO_Biological_process = []

for i in urls:
    if i==0:
        GO_Biological_process.append("Na")
    else:
        uclient=urlopen(i)
        pagehtml=uclient.read()
        uclient.close()                                    
        pagesoup=soup(pagehtml, "html.parser")
        page=pagesoup.find("ul",{"class":"noNumbering biological_process"}) 
        
        if page==None:
            GO_Biological_process.append("Na")
        else:
            location=page.find_all("a",{"onclick":"window.ga('UniProt-Entry-View', 'click', 'Display-GO-Term');"})
            #print(location)
            cadena = ""
            for i in range(0,len(location)):
                if location[i].text=="":
                    cadena = cadena + "-"
                else:
                    #GO_Biological_process.append(location[i].text)
                    cadena = cadena + location[i].text + "."
                #print(type(location[i].text))

            GO_Biological_process.append(cadena)

print(GO_Biological_process)
len(GO_Biological_process)
```
